# $G(n,p)$

This notebook aims to empirically investigate the properties of Erdős-Rényi graphs, specifically the $G(n,p)$ model, and compare them with real-world networks. Networkx provides a convenient way to generate random graphs using built-in functions for various commonly used graph models. To create an Erdős-Rényi graph, we can use the function `nx.erdos_renyi_graph(n, p)`, where `n` is the number of nodes and `p` is the probability of an edge between two nodes. The function returns an `nx.Graph` object containing the nodes and edges of the generated graph.


## Statistics of $G(n,p)$ (20 pts)

In the upcoming code cell, you will be requested to generate graphs from $G(n,p)$ model with $n$ set to 100, using different values of $p$ ranging from 0.001 to 0.081, with a small step size such as 0.005. For each value of $p$, you should generate 10 graph samples from the model and report the average of the following graph statistics with error bars for std. We also provide a brief explanation on how to compute these statistics using networkx functions:

- Number of nodes: To obtain the number of nodes in a given nx.graph object G, you can utilize the function `G.number_of_nodes()`.

- Number of edges: To obtain the number of edges in a given nx.graph object G, you can utilize the function `G.number_of_edges()`.

- Number of triangles: Using `nx.triangles(G)`, a dictionary of (node id, number of triangles participated) key-value pairs is returned. To calculate the total number of triangles in G, simply sum all the values in the dictionary and divide the result by 3 (since each triangle is counted three times in the dictionary).

- Number of isolated nodes: If a node has degree 0, then it is an isolated node. The function G.degree is a map-like object consisting of (node id, node degree) pairs. To count the number of isolated nodes, iterate through `G.degree` and count the number of 0s in the values.

- Number of connected components: In graph theory, a connected component is a set of vertices in a graph that are linked to each other by paths. Using nx.connected_components(G), a node list generator is returned, which yields one component at a time. To get the total number of connected components, you can use `len(list(nx.connected_components(G)))`.

For each graph statistic, generate a plot where the x-axis represents the values of $p$, and the y-axis represents the average of the statistic. 

```python
# Generate range of p values
p_range = [0.001 * (1 + x) for x in range(81)]

# Generate associated graphs
n = 100
sample_size = 10
G_range = [[nx.erdos_renyi_graph(n, p) for _ in range(sample_size)] for p in p_range]
```

```python
# nx.draw(G_range[0][5], node_size=40)
nx.draw(G_range[40][3], node_size=40)
# nx.draw(G_range[80][5], node_size=40)
```

```python
# Generate stats for number of edges.
edge_range = [[G.number_of_edges() for G in G_set] for G_set in G_range]
edge_means = [mean(edge_set) for edge_set in edge_range]
edge_stdev = [np.std(edge_set) for edge_set in edge_range]
```

```python
# plot with error bars.
fig, ax = plt.subplots()

plt.errorbar(p_range, edge_means, yerr=edge_stdev, label='n = 100')
plt.xlabel('Edge Formation Probability (p)')
plt.ylabel('Edge Count')

ax.set_title('Edge Count of G(100, p)')

plt.show()
```

```python
# Generate stats for number of triangles.
tri_range = [[int(sum(dict.values(nx.triangles(G)))/3) for G in G_set] for G_set in G_range]
tri_means = [mean(tri_set) for tri_set in tri_range]
tri_stdev = [np.std(tri_set) for tri_set in tri_range]
```

```python
# plot with error bars.
fig, ax = plt.subplots()

plt.errorbar(p_range, tri_means, yerr=tri_stdev, label='n = 100')
plt.xlabel('Edge Formation Probability (p)')
plt.ylabel('Triangle Count')

ax.set_title('Triangles in G(100, p)')

plt.show()
```

```python
# Generate stats for number of isolated nodes.
iso_range = [[nx.number_of_isolates(G) for G in G_set] for G_set in G_range]
iso_means = [mean(iso_set) for iso_set in iso_range]
iso_stdev = [np.std(iso_set) for iso_set in iso_range]
```

```python
# plot with error bars.
fig, ax = plt.subplots()

plt.errorbar(p_range, iso_means, yerr=iso_stdev, label='n = 100')
plt.xlabel('Edge Formation Probability (p)')
plt.ylabel('Isolated Nodes')

ax.set_title('Isolated Nodes in G(100, p)')

plt.show()
```

```python
# Generate stats for number of connected components.
compo_range = [[nx.number_connected_components(G) for G in G_set] for G_set in G_range]
compo_means = [mean(compo_set) for compo_set in compo_range]
compo_stdev = [np.std(compo_set) for compo_set in compo_range]
```

```python
# plot with error bars.
fig, ax = plt.subplots()

plt.errorbar(p_range, compo_means, yerr=compo_stdev, label='n = 100')
plt.xlabel('Edge Formation Probability (p)')
plt.ylabel('Connected Component Count')

ax.set_title('Connected Components in G(100, p)')

plt.show()
```

# Real world network

## Graph statistics (10pts)

In the same folder we provide a real-world social network dataset saved as edgelist format named "fb-pages-food.edges". Each line in this file has the format of "node1,node2" that represents an edge connecting node1 and node2. An edgelist file can be loaded as nx.Graph directly using nx.read_edgelist() like the following.

```python
FB = nx.read_edgelist("fb-pages-food.edges", delimiter=',')
print(FB)
nx.draw(FB, nx.spring_layout(FB,scale=1), node_size=10)
```

How many nodes and edges are there in this graph? Also report the number of triangles, number of isolated nodes, and number of connected components in this graph.

```python
# Report the graph statistics.
edge_FB = FB.number_of_edges()
tri_FB = int(sum(dict.values(nx.triangles(FB)))/3)
iso_FB = nx.number_of_isolates(FB)
compo_FB = nx.number_connected_components(FB)

print('Edges:\t\t\t', edge_FB, '\nTriangles:\t\t', tri_FB, 
      '\nIsolated Nodes:\t\t', iso_FB, '\nConnected Components:\t', compo_FB)
```

In the next cell, plot a histogram of the node degrees in this graph with bins=15.

```python
degrees = [val for (node, val) in FB.degree()]

fig, ax = plt.subplots()

plt.hist(degrees, bins=15)

ax.set_title('Node Degrees in FB (bins=15)')
plt.show()
```

## Try to fit the data with G(n,p) (10pts)

If we want to fit the graph with G(n,p) model, i.e., find a G(n,p) model whose expected number of edges equals to number of edge in this graph, How should we set p to get m edges in expectation? 

### CALCULATION:

The number of nodes should remain the same, so n = 620. The expected number of edges will be ${n}\choose{2}$ $\times$ $p$, so if we want to a graph with 2102 edges, then we have $p = \frac{2102}{620\choose2}$, which in turn evaluates to $\frac{2102}{191890} = 0.0109541925061233$

```python
p = 2102/191890
print('p =', p)
```

Now use the p value you calculated and sample 10 graphs from G(n,p). Report the same graph statistics in average.

```python
fit1 = [nx.erdos_renyi_graph(620, p) for _ in range(10)]

e_fit1 = np.mean([G.number_of_edges() for G in fit1])
t_fit1 = np.mean([int(sum(dict.values(nx.triangles(G)))/3) for G in fit1])
i_fit1 = np.mean([nx.number_of_isolates(G) for G in fit1])
c_fit1 = np.mean([nx.number_connected_components(G) for G in fit1])

print('Edges:\t\t\t', e_fit1, '\nTriangles:\t\t', t_fit1, 
      '\nIsolated Nodes:\t\t', i_fit1, '\nConnected Components:\t', c_fit1)
```

Plot the histogram of node degrees for one random graph you generated with bins=15, what's the difference between the node degree sequences of the real graph and the random graph?

```python
graphInQuestion = fit1[6]
deg_fit1 = [val for (node, val) in graphInQuestion.degree()]

fig, ax = plt.subplots()

plt.hist(deg_fit1, bins=15)

ax.set_title('Node Degrees in G(620, p) (bins=15)')
plt.show()
```

The primary difference between the provided FB graph and G(620, 0.0109541925061233) is that many nodes in FB had degrees above 20, with a few on the order of 40-60 (and one with degree 134!). However, there were no degrees above 16 in the Erdős–Rényi simulation. This means that in the FB graph, some nodes should be expected to have a higher degree than predicted!

## A random graph model that fits the degree sequence (10pts)

Given a sequence of expected degrees $(d_1,d_2\dots d_n)$ of length n, we generate a graph with n nodes, and assigns an edge between node $u$ and node $v$ with probability $$p_{uv}={d_ud_v\over \sum_k d_k}$$.

This model is known as the Chung-Lu model and is implemented in the networkx library. To generate a graph from this model, simply use the function `nx.expected_degree_graph(node_degree_list, selfloops=False)`.

To compare the generated graph with a real social network, pass the degree sequence of the real network into this model and generate 10 samples. Then plot the degree histogram of the generated graphs and compare it with the original graph. Additionally, report the same graph statistics for the generated graphs and compare them with the original graph. What observations can we make from this comparison?

```python
fit2 = [nx.expected_degree_graph(sorted(degrees), selfloops=False) for _ in range(10)]
deg_fit2 = [deg for (_, deg) in G.degree() for G in fit2]

fig, ax = plt.subplots()

plt.hist(deg_fit2, bins=15)

ax.set_title('Sum of Node Degrees in 10 Chung-Lu variates (bins=15)')
plt.show()
```

```python
# Average stats:
edge_fit2 = np.mean([G.number_of_edges() for G in fit2])
tri_fit2 = np.mean([int(sum(dict.values(nx.triangles(G)))/3) for G in fit2])
iso_fit2 = np.mean([nx.number_of_isolates(G) for G in fit2])
compo_fit2 = np.mean([nx.number_connected_components(G) for G in fit2])

print('Edges:\t\t\t', edge_fit2, '\nTriangles:\t\t', tri_fit2, 
      '\nIsolated Nodes:\t\t', iso_fit2, '\nConnected Components:\t', compo_fit2)
```

Using the Chung-Lu model, our graphs on average contained more isolated nodes and distinct connected components than either the FB graph or the G(n,p) simulation. However, the number of triangles was much closer to the expected number, though still far lower than observed in the FB model (cliques seem far less likely in simulations than they do in the actual FB graph). In the FB graph, there were no isolated nodes and a single connected component, which may have something to do with how the data was harvested, but could also be indicative of tha fact that real-world networks have higher levels of connectivity with the same node density.
